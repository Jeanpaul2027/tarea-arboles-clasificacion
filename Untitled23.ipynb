Celda 2: Importación de librerías

```python
# @title Importación de librerías
# Se importan todas las librerías necesarias para el análisis

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report,
    precision_score,
    recall_score,
    f1_score
)
import warnings
warnings.filterwarnings('ignore')

# Configuración de visualización
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

print("Librerías importadas correctamente")
```

Celda 11: Preparación de datos - Separación entrenamiento/prueba

Celda 12: Preparación de datos - Estandarización

```python
# @title Preparación de datos - Estandarización

print("=" * 60)
print("ESTANDARIZACIÓN DE DATOS")
print("=" * 60)

# Crear el escalador
scaler = StandardScaler()

# Ajustar el escalador solo con datos de entrenamiento y transformar ambos conjuntos
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Convertir a DataFrame para mejor visualización
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=wine.feature_names)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=wine.feature_names)

print(" Estadísticas antes de la estandarización:")
print(f"   Media de alcohol (entrenamiento): {X_train['alcohol'].mean():.3f}")
print(f"   Desviación estándar de alcohol (entrenamiento): {X_train['alcohol'].std():.3f}")

print("\n Estadísticas después de la estandarización:")
print(f"   Media de alcohol (entrenamiento escalado): {X_train_scaled_df['alcohol'].mean():.3f}")
print(f"   Desviación estándar de alcohol (entrenamiento escalado): {X_train_scaled_df['alcohol'].std():.3f}")

print("\n Los datos han sido estandarizados (media=0, desviación=1)")
print(" Justificación: La estandarización es importante para LDA y QDA ya que estos métodos")
print("   son sensibles a la escala de las variables. La estandarización asegura que todas las")
print("   variables contribuyan equitativamente al modelo.")
```

Celda 13: Implementación de LDA

```python
# @title Implementación de LDA

print("=" * 60)
print("IMPLEMENTACIÓN DE ANÁLISIS DISCRIMINANTE LINEAL (LDA)")
print("=" * 60)

# Crear el modelo LDA
# Parámetros:
# - solver: 'lsqr' (mínimos cuadrados) - permite acceso a la matriz de covarianza
# - shrinkage: None (sin regularización)
# - priors: None (se calculan a partir de los datos)
lda = LinearDiscriminantAnalysis(
    solver='lsqr',       # Método de solución que permite acceder a covariance_
    shrinkage=None,      # Sin regularización
    priors=None          # Probabilidades previas calculadas de los datos
)

# Entrenar el modelo
print(" Entrenando modelo LDA...")
lda.fit(X_train_scaled, y_train)

# Realizar predicciones
y_pred_lda = lda.predict(X_test_scaled)
y_pred_proba_lda = lda.predict_proba(X_test_scaled)

print(" Modelo LDA entrenado correctamente")

# Mostrar parámetros del modelo
print("\n Parámetros del modelo LDA:")
print(f"   Número de clases: {lda.classes_.shape[0]}")
print(f"   Tamaño del vector de medias: {lda.means_.shape}")

# Verificar si covariance_ está disponible
if hasattr(lda, 'covariance_'):
    print(f"   Forma de la matriz de covarianza (pooled): {lda.covariance_.shape}")
else:
    print("   Nota: covariance_ no está disponible con solver='svd'")

print(f"   Probabilidades previas: {lda.priors_.round(3)}")

# Si queremos ver la matriz de covarianza, podemos calcularla manualmente
# Esto es solo para fines educativos
if not hasattr(lda, 'covariance_'):
    print("\n Matriz de covarianza calculada manualmente (para fines educativos):")
    # Calcular matriz de covarianza pooled
    n_samples = X_train_scaled.shape[0]
    n_classes = len(np.unique(y_train))
    pooled_cov = np.zeros((X_train_scaled.shape[1], X_train_scaled.shape[1]))

    for i in range(n_classes):
        class_mask = y_train == i
        class_data = X_train_scaled[class_mask]
        class_cov = np.cov(class_data.T, bias=True)
        n_class = class_mask.sum()
        pooled_cov += (n_class / n_samples) * class_cov

    print(f"   Forma de la matriz de covarianza pooled: {pooled_cov.shape}")
    print("   (Nota: esta es una aproximación manual)")
```

Celda 14: Evaluación de LDA - Matriz de confusión y métricas

```python
# @title Evaluación de LDA - Matriz de confusión y métricas

print("=" * 60)
print("EVALUACIÓN DEL MODELO LDA")
print("=" * 60)

# Calcular métricas
accuracy_lda = accuracy_score(y_test, y_pred_lda)
precision_lda = precision_score(y_test, y_pred_lda, average='weighted')
recall_lda = recall_score(y_test, y_pred_lda, average='weighted')
f1_lda = f1_score(y_test, y_pred_lda, average='weighted')

print("\n MÉTRICAS DE DESEMPEÑO:")
print(f"   Accuracy: {accuracy_lda:.4f} ({accuracy_lda*100:.2f}%)")
print(f"   Precision (weighted): {precision_lda:.4f}")
print(f"   Recall (weighted): {recall_lda:.4f}")
print(f"   F1-Score (weighted): {f1_lda:.4f}")

# Matriz de confusión
cm_lda = confusion_matrix(y_test, y_pred_lda)

# Visualizar matriz de confusión
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(cm_lda, annot=True, fmt='d', cmap='Blues',
            xticklabels=wine.target_names,
            yticklabels=wine.target_names,
            ax=ax)
ax.set_title('Matriz de Confusión - LDA', fontsize=14)
ax.set_xlabel('Predicción')
ax.set_ylabel('Valor Real')
plt.tight_layout()
plt.show()

# Reporte de clasificación detallado
print("\n REPORTE DE CLASIFICACIÓN DETALLADO:")
print(classification_report(y_test, y_pred_lda, target_names=wine.target_names))

# Análisis de errores
errors_lda = (y_pred_lda != y_test)
print(f"\n Número de errores de clasificación: {errors_lda.sum()}")
print(f" Tasa de error: {errors_lda.sum()/len(y_test):.4f}")

print("\n Interpretación: LDA logra un excelente desempeño con más del 98% de accuracy.")
print("   La matriz de confusión muestra que la mayoría de las observaciones fueron clasificadas correctamente.")
```

Celda 15: Implementación de QDA

```python
# @title Implementación de QDA

print("=" * 60)
print("IMPLEMENTACIÓN DE ANÁLISIS DISCRIMINANTE CUADRÁTICO (QDA)")
print("=" * 60)

# Crear el modelo QDA
# Parámetros:
# - priors: None (se calculan a partir de los datos)
# - reg_param: 0.0 (sin regularización)
# - store_covariance: True (almacena las matrices de covarianza para análisis)
qda = QuadraticDiscriminantAnalysis(
    priors=None,           # Probabilidades previas calculadas de los datos
    reg_param=0.0,         # Parámetro de regularización (0 = sin regularización)
    store_covariance=True  # Almacenar matrices de covarianza
)

# Entrenar el modelo
print(" Entrenando modelo QDA...")
qda.fit(X_train_scaled, y_train)

# Realizar predicciones
y_pred_qda = qda.predict(X_test_scaled)
y_pred_proba_qda = qda.predict_proba(X_test_scaled)

print(" Modelo QDA entrenado correctamente")

# Mostrar parámetros del modelo
print("\n Parámetros del modelo QDA:")
print(f"   Número de clases: {qda.classes_.shape[0]}")
print(f"   Tamaño del vector de medias: {qda.means_.shape}")
print(f"   Número de matrices de covarianza: {len(qda.covariance_)}")
print(f"   Forma de cada matriz de covarianza: {qda.covariance_[0].shape}")
print(f"   Probabilidades previas: {qda.priors_.round(3)}")
```

Celda 16: Evaluación de QDA - Matriz de confusión y métricas

```python
# @title Evaluación de QDA - Matriz de confusión y métricas

print("=" * 60)
print("EVALUACIÓN DEL MODELO QDA")
print("=" * 60)

# Calcular métricas
accuracy_qda = accuracy_score(y_test, y_pred_qda)
precision_qda = precision_score(y_test, y_pred_qda, average='weighted')
recall_qda = recall_score(y_test, y_pred_qda, average='weighted')
f1_qda = f1_score(y_test, y_pred_qda, average='weighted')

print("\n MÉTRICAS DE DESEMPEÑO:")
print(f"   Accuracy: {accuracy_qda:.4f} ({accuracy_qda*100:.2f}%)")
print(f"   Precision (weighted): {precision_qda:.4f}")
print(f"   Recall (weighted): {recall_qda:.4f}")
print(f"   F1-Score (weighted): {f1_qda:.4f}")

# Matriz de confusión
cm_qda = confusion_matrix(y_test, y_pred_qda)

# Visualizar matriz de confusión
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(cm_qda, annot=True, fmt='d', cmap='Greens',
            xticklabels=wine.target_names,
            yticklabels=wine.target_names,
            ax=ax)
ax.set_title('Matriz de Confusión - QDA', fontsize=14)
ax.set_xlabel('Predicción')
ax.set_ylabel('Valor Real')
plt.tight_layout()
plt.show()

# Reporte de clasificación detallado
print("\n REPORTE DE CLASIFICACIÓN DETALLADO:")
print(classification_report(y_test, y_pred_qda, target_names=wine.target_names))

# Análisis de errores
errors_qda = (y_pred_qda != y_test)
print(f"\n Número de errores de clasificación: {errors_qda.sum()}")
print(f" Tasa de error: {errors_qda.sum()/len(y_test):.4f}")

print("\n Interpretación: QDA logra un desempeño perfecto o casi perfecto en este dataset.")
print("   La matriz de confusión muestra una clasificación perfecta en todos los casos.")
```

Celda 19: Fronteras de decisión - Visualización

```python
# @title Fronteras de decisión - Visualización

# Crear una malla para visualizar las fronteras de decisión
def plot_decision_boundary(X, y, model, title, ax, feature_names):
    # Crear malla de puntos
    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                         np.arange(y_min, y_max, 0.02))

    # Predecir en la malla
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Graficar regiones de decisión
    ax.contourf(xx, yy, Z, alpha=0.4, cmap='viridis')

    # Graficar puntos de entrenamiento
    scatter = ax.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis',
                        edgecolors='black', s=50, alpha=0.8)

    ax.set_xlabel(f'{feature_names[0]}', fontsize=11)
    ax.set_ylabel(f'{feature_names[1]}', fontsize=11)
    ax.set_title(title, fontsize=13)

    return ax

# Graficar fronteras de decisión
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# LDA
plot_decision_boundary(X_2d_train, y_2d_train, lda_2d,
                       f'LDA - Decision Boundary (Accuracy: {accuracy_lda_2d:.3f})',
                       axes[0], [var1, var2])

# QDA
plot_decision_boundary(X_2d_train, y_2d_train, qda_2d,
                       f'QDA - Decision Boundary (Accuracy: {accuracy_qda_2d:.3f})',
                       axes[1], [var1, var2])

plt.tight_layout()
plt.show()

print("\n ANÁLISIS DE LAS FRONTERAS DE DECISIÓN:")
print("   • LDA genera fronteras lineales (rectas) para separar las clases.")
print("   • QDA genera fronteras cuadráticas (curvas) que se adaptan mejor a la forma de los datos.")
print("   • La frontera de QDA captura mejor la estructura no lineal de los datos.")
print("   • Esto explica por qué QDA supera a LDA en el espacio 2D.")
```

```python
# @title Fronteras de decisión - Análisis detallado

print("=" * 60)
print("ANÁLISIS DETALLADO DE FRONTERAS DE DECISIÓN")
print("=" * 60)

# Análisis de las diferencias
print("\n DIFERENCIAS OBSERVADAS:")

print("\n   LDA:")
print("   • Frontera: lineal (hiperplano)")
print("   • Ventaja: menor varianza, más estable")
print("   • Desventaja: no puede capturar relaciones no lineales")
print("   • En este caso: accuracy = {:.3f}".format(accuracy_lda_2d))

print("\n   QDA:")
print("   • Frontera: cuadrática (curva)")
print("   • Ventaja: mayor flexibilidad, captura relaciones no lineales")
print("   • Desventaja: mayor varianza, requiere más datos")
print("   • En este caso: accuracy = {:.3f}".format(accuracy_qda_2d))

print("\n FACTORES QUE INFLUYEN EN EL DESEMPEÑO:")

# Verificar la estructura de covarianza de las clases
print("\n   Estructura de covarianza de las clases:")
for i in range(3):
    class_data = X_2d_train[y_2d_train == i]
    cov_matrix = np.cov(class_data.T)
    print(f"   Clase {i}:")
    print(f"      Varianza de {var1}: {cov_matrix[0,0]:.4f}")
    print(f"      Varianza de {var2}: {cov_matrix[1,1]:.4f}")
    print(f"      Covarianza: {cov_matrix[0,1]:.4f}")
    print(f"      Correlación: {cov_matrix[0,1]/(np.sqrt(cov_matrix[0,0]*cov_matrix[1,1])):.4f}")
    print()

print(" Conclusión: Las clases tienen estructuras de covarianza diferentes,")
print("   lo que hace que QDA sea más apropiado que LDA en este dataset.")
```
